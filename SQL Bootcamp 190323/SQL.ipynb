# SQL
SQL (Structured Query Language) is a standard programming language for managing and manipulating relational databases. It is used to insert, update, and retrieve data from a database, as well as to manage database structure and user access control. SQL is supported by many relational database management systems including MySQL, PostgreSQL, and Microsoft SQL Server.

# RDMS
A Relational Database Management System (RDMS) is a software system that is used to manage and organize data stored in relational databases. It allows you to store and retrieve data by creating tables, defining relationships between tables, and using SQL to manipulate and query the data.

For example, consider a simple database for a library. The database has three tables: Books, Members, and Loans. The Books table stores information about the books in the library such as title, author, and ISBN. The Members table stores information about library members such as name, address, and membership number. The Loans table stores information about which books have been loaned to which members and when they are due to be returned.

By using relationships between the tables (e.g., a member can have many loans, but a book can only be loaned to one member at a time), the RDMS can help you to keep track of which books have been loaned, who they have been loaned to, and when they are due to be returned. This makes it easy to search for information, update records, and generate reports.

SQL is the most common language used to interact with relational databases.

The AND operator is used to combine two or more conditions in a WHERE clause. It returns true if all of the conditions are true, and false if any of the conditions are false.

## OR Operator: 
The OR operator is used to combine two or more conditions in a WHERE clause. It returns true if any of the conditions are true, and false if all of the conditions are false.


For example, 
to retrieve all rows from the employees table where 
- the birth_date column is equal to '1964-06-02' OR 
- the first_name column is equal to "Lech" 

you can use the following SQL statement with the OR operator:

```python
%%sql

SELECT * FROM employees 
WHERE first_name = 'Lech' OR 
birth_date = '1964-06-02'

limit 10
;
```

## IN Operator: 
The IN operator is used to specify a list of values to match against a column. It returns true if the value matches any of the values in the list, and false otherwise.

For example, 
to retrieve all rows from the employees table where 
- the first_name column is either "Bezalel" or "Lech"

you can use the following SQL statement with the IN operator:

```python
%%sql

SELECT * FROM employees 
WHERE first_name in ('Lech','Bezalel')

limit 10
;
```

## LIKE Operator: 
The LIKE operator is used to match a pattern against a string column. It returns true if the pattern matches any part of the string, and false otherwise.


For example, 
to retrieve all rows from the employees table where 
- the first_name column starts with "J", 


you can use the following SQL statement with the LIKE operator and wildcard character '%':

```python
%%sql

SELECT * FROM employees WHERE first_name LIKE 'J%'

limit 10
;
```

## Wildcard Operator: 
The Wildcard operators are used to match patterns in string values. There are two wildcard operators in SQL:


- % (percent sign): Matches any string of zero or more characters.
- _ (underscore): Matches any single character.


For example, to retrieve all rows from the employees table where 
- the last_name column ends with "son", 

you can use the following SQL statement with the LIKE operator and the % wildcard character:

```python
%%sql

SELECT * FROM employees WHERE last_name LIKE '%son'

limit 10
;
```

## GROUP BY: 
The GROUP BY statement is used to group the result set by one or more columns. It is used with aggregate functions to group the data and calculate summary statistics.

For example, to group the employees table by the gender column and count the number of employees in each group, you can use the following SQL statement:

```python
%%sql

SELECT gender, COUNT(*) FROM employees 
GROUP BY gender

limit 10
;
```

## HAVING: 
The HAVING statement is used to filter the results of a GROUP BY query based on a specified condition. It is used to extract only the groups that meet the condition.

For example, to group the **`employees`** table by the **`gender`** column and count the number of **`employees`** in each group, but only return groups with more than 100 employees, you can use the following SQL statement:

```python
%%sql

SELECT gender, COUNT(*) FROM employees 
GROUP BY gender 
HAVING COUNT(*) > 100

limit 10
;
```

## ORDER BY: 
The ORDER BY statement is used to sort the result set by one or more columns in ascending or descending order.

For example, to retrieve all columns from the **`employees`** table and sort them by the **`last_name`** column in ascending order, you can use the following SQL statement:

```python
%%sql

SELECT * FROM employees 
ORDER BY last_name ASC

limit 10
;
```

## INNER JOIN
The INNER JOIN is used to return only the rows in both tables that satisfy the join condition. For example, to select all students from the "student" table and their corresponding scores from the "score" table, you can use the following SQL code: 

```python
%%sql
SELECT employees.emp_no, employees.first_name, employees.last_name, titles.title
FROM employees
INNER JOIN titles
ON employees.emp_no = titles.emp_no
limit 10;
```

## LEFT JOIN
The LEFT JOIN returns all the rows from the left table (employees), along with the matching rows from the right table (titles). If there are no matching rows in the right table, it returns NULL. Here's an example of a LEFT JOIN between the employees and titles tables on emp_no:

```python
%%sql
SELECT E.*, d.dept_no, d.from_date, d.to_date
FROM employees.employees e
LEFT JOIN employees.dept_manager d
ON e.emp_no = d.emp_no
# where dept_no is not null
LIMIT 10;
```

```python
%%sql
SELECT E.*, d.dept_no, d.from_date, d.to_date
FROM employees.employees e
LEFT JOIN employees.dept_manager d
ON e.emp_no = d.emp_no
where e.emp_no in (110022,10003,110386,110228,10007,110085,10009)
LIMIT 100;
```

## RIGHT JOIN
The RIGHT JOIN returns all the rows from the right table (titles), along with the matching rows from the left table (employees). If there are no matching rows in the left table, it returns NULL. Here's an example of a RIGHT JOIN between the employees and titles tables on emp_no:

```python
%%sql
SELECT E.*, d.dept_no, d.from_date, d.to_date
FROM employees.dept_manager d
RIGHT JOIN employees.employees e
ON e.emp_no = d.emp_no
where e.emp_no in (110022,10003,110386,110228,10007,110085,10009)
LIMIT 100;
```

## OUTER JOIN
An OUTER JOIN returns all the rows from both tables, along with the matching rows based on the join condition. If there is no match, NULL is returned for the missing values

```python
%%sql
SELECT employees.emp_no, employees.first_name, employees.last_name, titles.title
FROM employees
OUTER JOIN titles
ON employees.emp_no = titles.emp_no
LIMIT 10;
```

## CASE WHEN
The CASE WHEN statement is a conditional expression that returns a value based on a condition. It is used to perform IF-THEN-ELSE logic in SQL. The syntax of the CASE WHEN statement is:

```python
# SELECT
#     column_name,
#     CASE
#         WHEN condition1 THEN result1
#         WHEN condition2 THEN result2
#         ELSE result3
#     END
# FROM table_name;
```

Here's an example of using the CASE WHEN statement to create a new column in the employees table based on the gender of the employee:

```python
%%sql
SELECT emp_no, first_name, last_name, gender,
    CASE
        WHEN gender = 'M' THEN 'Male'
        WHEN gender = 'F' THEN 'Female'
        ELSE 'Unknown'
    END AS gender_full
FROM employees
LIMIT 10;
```

## Window Functions
Window Functions: Window functions are a type of SQL function that perform calculations across a set of rows that are related to the current row. They are used to compute values that are based on a subset of rows in the result set. Some common window functions include RANK, ROW_NUMBER, and DENSE_RANK.
- **`ROW_NUMBER:`** The ROW_NUMBER function assigns a unique number to each row in the result set. The numbering starts at 1 for the first row and increments by 1 for each subsequent row. Here's an example of using the ROW_NUMBER function to rank employees based on their hire date:

```python
%%sql
SELECT emp_no, first_name, last_name, hire_date,
       ROW_NUMBER() OVER(ORDER BY hire_date) AS hire_rank
FROM employees
LIMIT 10;
```

- **`RANK:`** The RANK function assigns a rank to each row in the result set based on the value of a specified column. If two or more rows have the same value, they are assigned the same rank, and the next rank is skipped. Here's an example of using the RANK function to rank employees based on their salary:

```python
%%sql
SELECT emp_no, salary,
       RANK() OVER(ORDER BY salary DESC) AS salary_rank
FROM salaries
LIMIT 10;
```

- **`DENSE_RANK:`** The DENSE_RANK function assigns a rank to each row in the result set based on the value of a specified column. If two or more rows have the same value, they are assigned the same rank, and the next rank is not skipped. Here's an example of using the DENSE_RANK function to rank employees based on their salary:

```python
%%sql
SELECT emp_no, salary,
       DENSE_RANK() OVER(ORDER BY salary DESC) AS salary_rank
FROM salaries
LIMIT 10;
```
